```python
# üß© Step 1: Install OpenCV
!pip install opencv-python --quiet

# üñºÔ∏è Step 2: Upload Image
from google.colab import files
uploaded = files.upload()

import cv2
import numpy as np
from matplotlib import pyplot as plt
from pathlib import Path

# Get the uploaded file name
input_image_path = next(iter(uploaded))

# üì∑ Step 3: Read and Convert to Grayscale
image = cv2.imread(input_image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# üíæ Step 4: Save the Grayscale Image
output_image_path = "grayscale_" + Path(input_image_path).name
cv2.imwrite(output_image_path, gray_image)

# üì• Step 5: Download the Image
files.download(output_image_path)

# üñºÔ∏è Step 6: Display Images (Original and Grayscale)
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Grayscale Image")
plt.imshow(gray_image, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
